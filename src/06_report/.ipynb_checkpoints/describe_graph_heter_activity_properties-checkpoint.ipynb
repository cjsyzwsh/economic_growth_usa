## Activity sorting

```python
# create activity counts
activity_counts = np.unique(df.cat, return_counts = True)

# modify it to dataframes
activity_counts_df = pd.DataFrame(activity_counts[1], 
                                  index = activity_counts[0], 
                                  columns = ['value'])

# print
pd.set_option('display.max_rows', 700)
activity_counts_df.sort_values('value')
```

```python
# return top 10
activity_counts_df.sort_values('value', ascending=False).iloc[:10, :]

# informative activities (from the regression results.)
econ_activity_list = ['Caribbean', 'Latin American', 'Fried Chicken', 'Spanish', 
                      'Brazilian', 'Poutine Place', 'Tattoo', 'Discount Store', 
                      'French', 'New American', 'Cycle Studio', 'Salad', 'Peking Duck', 
                      'Juice Bar', 'Art Museum', 'Climbing Gym']

# return the index
# activity_counts_df.sort_values('value', ascending=False).index.values

# print the ranking of the important economic activities
econ_activity_locations = activity_counts_df.sort_values('value', ascending=False).index.isin(econ_activity_list)
econ_activity_ranking = np.arange(activity_counts_df.shape[0])[econ_activity_locations]
print("Ranking of the important econ activities:", econ_activity_ranking)
```
