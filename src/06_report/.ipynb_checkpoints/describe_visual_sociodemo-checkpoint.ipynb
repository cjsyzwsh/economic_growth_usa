```python
# read files
with open('../data/02_intermediate/boston_shp.pickle','rb') as f: 
    boston_shp = pickle.load(f)

with open('../data/02_intermediate/boston_socioecon_2013.pickle', 'rb') as f:
    boston_socioecon_2013 = pickle.load(f)

with open('../data/02_intermediate/boston_socioecon_2016.pickle', 'rb') as f:
    boston_socioecon_2016 = pickle.load(f)
    
with open('../data/02_intermediate/boston_socioecon_2018.pickle', 'rb') as f:
    boston_socioecon_2018 = pickle.load(f)

with open('../data/02_intermediate/boston_annual_growth_2013_2018.pickle', 'rb') as f:
    annual_growth_2013_2018 = pickle.load(f)

with open('../data/02_intermediate/boston_annual_growth_2016_2018.pickle', 'rb') as f:
    annual_growth_2016_2018 = pickle.load(f)
    
```

```python
# save path
save_path = '../data/06_reporting/'
```

```python
# describe nine variables by hist.
# pop_total, rent_median, household_size_avg, race_white_ratio, race_black_ratio, 
# employment_unemployed_ratio, vehicle_per_capita, property_value_median, vacancy_ratio

df_dic = {'2013':boston_socioecon_2013, '2016':boston_socioecon_2016, '2018':boston_socioecon_2018}

for k in df_dic.keys():
    df = df_dic[k]
    print("Year ", k)

    fig, ax = plt.subplots(3,3, figsize = (20,20))

    ax[0,0].hist(df['pop_total'])
    avg = np.round(np.mean(df['pop_total']), decimals = 2)
    ax[0,0].set_title(str(k)+'Total population - average '+str(avg))

    ax[0,1].hist(df.loc[df['household_size_avg'] < 10, 'household_size_avg']) # outlier exists for average hh size (2013).
    avg = np.round(np.mean(df.loc[df['household_size_avg'] < 10, 'household_size_avg']), decimals = 2)
    ax[0,1].set_title(str(k)+'Average household size - average '+str(avg))

    ax[0,2].hist(df['race_white_ratio'])
    avg = np.round(np.mean(df['race_white_ratio']), decimals = 2)
    ax[0,2].set_title(str(k)+'white ratio - average '+str(avg))

    ax[1,0].hist(df['race_black_ratio'])
    avg = np.round(np.mean(df['race_black_ratio']), decimals = 2)
    ax[1,0].set_title(str(k)+'black ratio - average '+str(avg))

    ax[1,1].hist(df['rent_median'])
    avg = np.round(np.mean(df['rent_median']), decimals = 2)
    ax[1,1].set_title(str(k)+'median rent - average '+str(avg))

    ax[1,2].hist(df['employment_unemployed_ratio'])
    avg = np.round(np.mean(df['employment_unemployed_ratio']), decimals = 2)
    ax[1,2].set_title(str(k)+'unemployement rate - average'+str(avg))

    ax[2,0].hist(df['vehicle_per_capita'])
    avg = np.round(np.mean(df['vehicle_per_capita']), decimals = 2)
    ax[2,0].set_title(str(k)+'vehicle_per_capita - average'+ str(avg))

    ax[2,1].hist(df['property_value_median'])
    avg = np.round(np.mean(df['property_value_median']), decimals = 2)
    ax[2,1].set_title(str(k)+'median property value - average '+str(avg))

    ax[2,2].hist(df['vacancy_ratio'])
    avg = np.round(np.mean(df['vacancy_ratio']), decimals = 2)
    ax[2,2].set_title(str(k)+'vacancy rate - average' + str(avg))
    
    fig.savefig(save_path + 'socio_econ_hist_' + str(k) + '.png')
#     plt.close()
```

```python
# visualize growth
df_dic = {'2013-2018':annual_growth_2013_2018, '2016-2018':annual_growth_2016_2018}

for k in df_dic.keys():
    df = df_dic[k]
    print("Year ", k)

    fig, ax = plt.subplots(2,3, figsize = (20,10))

    ax[0,0].hist(df['inc_per_capita_annual_growth'])
    avg = np.round(np.mean(df['inc_per_capita_annual_growth']), decimals = 2) # income growth has outlier.
    ax[0,0].set_title(str(k)+'income growth - average '+str(avg))

    ax[0,1].hist(df['pop_total_annual_growth']) # outlier exists for average hh size (2013).
    avg = np.round(np.mean(df['pop_total_annual_growth']), decimals = 2)
    ax[0,1].set_title(str(k)+'population growth - average '+str(avg))

    ax[0,2].hist(df['property_value_median_annual_growth'])
    avg = np.round(np.mean(df['property_value_median_annual_growth']), decimals = 2)
    ax[0,2].set_title(str(k)+'property value growth - average '+str(avg))
    
    fig.savefig(save_path + 'socio_econ_growth_hist_' + str(k) + '.png')
#     plt.close()
    
```

```python
# prepare dictionary data
boston_socioecon_dic = {'2013':boston_socioecon_2013, '2016':boston_socioecon_2016, '2018':boston_socioecon_2018}
boston_growth_dic = {'2013_2018':annual_growth_2013_2018, '2016_2018':annual_growth_2016_2018}

col_list_static = ['pop_total', 'inc_per_capita', 'race_white_ratio', 'race_black_ratio', 'employment_unemployed_ratio',
                   'vacancy_ratio', 'property_value_median', 'travel_driving_ratio', 'edu_bachelor_ratio']
col_list_growth = ['inc_per_capita_annual_growth', 'pop_total_annual_growth', 
                   'property_value_median_annual_growth']
```

```python
# visualize each year (whole area)
alpha = 0.7

for year in boston_socioecon_dic.keys():
    boston_socioecon = boston_socioecon_dic[year]
    boston_socioecon_shp = pd.concat([boston_shp, boston_socioecon], axis = 1)
    
    # save_path = '../data/06_reporting/'
    for col in col_list_static:
        fig_name = year + '_' + col 
        utils.plot_node_with_base_map(boston_socioecon_shp, 
                                      col, # column name
                                      col, # graph name
                                      alpha, fig_name, save_path)
```
