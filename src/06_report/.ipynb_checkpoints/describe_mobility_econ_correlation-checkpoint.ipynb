# Processing

```python
### turn the dictionary of adjacency matrices to data frames

###
A_unweighted_features_dic = {}
name_list = []

# A_unweighted.
for key_ in A_unweighted_dic.keys():
    A = A_unweighted_dic[key_]
    A_unweighted_features_dic[key_]=pd.DataFrame(A_unweighted_dic[key_].mean(axis = 1)).T # mean
    name_list.append(key_)
# 
A_unweighted_features = pd.concat(A_unweighted_features_dic).T
A_unweighted_features.columns = name_list
A_unweighted_features.index = X_place.index

    
###
A_weighted_features_dic = {}
name_list = []
        
# A_weighted.
for key_ in A_weighted_dic.keys():
    A = A_weighted_dic[key_]
    A_weighted_features_dic[key_]=pd.DataFrame(A_weighted_dic[key_].mean(axis = 1)).T # mean
    name_list.append(key_)

# 
A_weighted_features = pd.concat(A_weighted_features_dic).T
A_weighted_features.columns = name_list
A_weighted_features.index = X_place.index
```

```python
# Create and check the R2 matrix
# SW
# (1) ratio of driving and public transit has quite high R2. 
# (2) total population and households have fine R2. 
# (3) growth prediction r2 is really really low - same as the previous results. 
# (4) disaggregation of activity counts seems to help

r2_matrix_full = pd.DataFrame(np.zeros((len(var_list_1 + ['all_counts']), len(var_list_2))), 
                         index = var_list_1 + ['all_counts'], columns = var_list_2)
r2_matrix_no_max = pd.DataFrame(np.zeros((len(var_list_1 + ['all_counts']), len(var_list_2))), 
                         index = var_list_1 + ['all_counts'], columns = var_list_2)
r2_matrix_no_max_zeros = pd.DataFrame(np.zeros((len(var_list_1 + ['all_counts']), len(var_list_2))), 
                         index = var_list_1 + ['all_counts'], columns = var_list_2)

for var_1 in var_list_1 + ['all_counts']:    
    for var_2 in var_list_2:
        if var_1 == 'all_counts': # Use all the count columns
            # 1. full data set
            X = df_shp_w_activities_socioecon[var_list_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_full.loc[var_1, var_2] = r2
            
            # 2. data set without top K activities
            X = df_shp_w_activities_socioecon_without_max[var_list_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon_without_max[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_no_max.loc[var_1, var_2] = r2
            
            # 3. data set without top K activities and zero values
            X = df_shp_w_activities_socioecon_without_max_zeros[var_list_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon_without_max_zeros[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_no_max_zeros.loc[var_1, var_2] = r2
            
            
        else: # Use only individual count columns
            # 1. full data set
            X = df_shp_w_activities_socioecon[var_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_full.loc[var_1, var_2] = r2
            
            # 2. data set without top K activities
            X = df_shp_w_activities_socioecon_without_max[var_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon_without_max[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_no_max.loc[var_1, var_2] = r2
            
            # 3. data set without top K activities and zero values
            X = df_shp_w_activities_socioecon_without_max_zeros[var_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon_without_max_zeros[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_no_max_zeros.loc[var_1, var_2] = r2
```

## Test the effects in log transformation

```python
# create the log transformation
# SW: The log transformation cannot be directly applied to the annual growth numbers since many are negative! 
df_shp_w_activities_socioecon_log = copy.copy(df_shp_w_activities_socioecon)
df_shp_w_activities_socioecon_without_max_log = copy.copy(df_shp_w_activities_socioecon_without_max)
df_shp_w_activities_socioecon_without_max_zeros_log = copy.copy(df_shp_w_activities_socioecon_without_max_zeros)
```

```python
# create the list of variables. 
var_list_1 = ['weighted_counts', 'unweighted_counts', 'weekday_weighted', 'weekend_weighted', 
            'hours_morning_rush_weighted', 'hours_afternoon_rush_weighted', 'hours_noon_weighted', 
            'hours_early_morning_weighted', 'hours_late_night_weighted', 'activity_office_weighted', 
            'activity_residence_weighted', 'activity_others_weighted', 'duration_less_1hour_weighted',
            'duration_between_1hour_3hour_weighted','duration_large_3hour_weighted']

var_list_2 = ['pop_total', 'households', 'inc_per_capita', 'employment_unemployed_ratio', 'travel_driving_ratio', 
            'travel_pt_ratio', 'vehicle_per_capita', 'property_value_median'] #, 
            # 'inc_per_capita_annual_growth', 'pop_total_annual_growth', 'property_value_median_annual_growth']

var_list = var_list_1 + var_list_2
```

```python
# lift the numbers by epsilon
epsilon = 0.01
df_shp_w_activities_socioecon_log[var_list] += epsilon
df_shp_w_activities_socioecon_without_max_log[var_list] += epsilon
df_shp_w_activities_socioecon_without_max_zeros_log[var_list] += epsilon

# use the log transformation
df_shp_w_activities_socioecon_log[var_list] = np.log(df_shp_w_activities_socioecon_log[var_list])
df_shp_w_activities_socioecon_without_max_log[var_list] = np.log(df_shp_w_activities_socioecon_without_max_log[var_list])
df_shp_w_activities_socioecon_without_max_zeros_log[var_list] = np.log(df_shp_w_activities_socioecon_without_max_zeros_log[var_list])
```

```python
# Create and check the R2 matrix in the log space
# SW findings
# (1) ratio of driving and public transit has quite high R2. 
# (2) total population and households have fine R2. 
# (3) growth prediction r2 is really really low - same as the previous results. 
# (4) disaggregation of activity counts seems to help

r2_matrix_full_log = pd.DataFrame(np.zeros((len(var_list_1 + ['all_counts']), len(var_list_2))), 
                         index = var_list_1 + ['all_counts'], columns = var_list_2)
r2_matrix_no_max_log = pd.DataFrame(np.zeros((len(var_list_1 + ['all_counts']), len(var_list_2))), 
                         index = var_list_1 + ['all_counts'], columns = var_list_2)
r2_matrix_no_max_zeros_log = pd.DataFrame(np.zeros((len(var_list_1 + ['all_counts']), len(var_list_2))), 
                         index = var_list_1 + ['all_counts'], columns = var_list_2)

for var_1 in var_list_1 + ['all_counts']:    
    for var_2 in var_list_2:
        if var_1 == 'all_counts': # Use all the count columns
            # 1. full data set
            X = df_shp_w_activities_socioecon_log[var_list_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon_log[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_full_log.loc[var_1, var_2] = r2
            
            # 2. data set without top K activities
            X = df_shp_w_activities_socioecon_without_max_log[var_list_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon_without_max_log[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_no_max_log.loc[var_1, var_2] = r2
            
            # 3. data set without top K activities and zero values
            X = df_shp_w_activities_socioecon_without_max_zeros_log[var_list_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon_without_max_zeros_log[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_no_max_zeros_log.loc[var_1, var_2] = r2
            
            
        else: # Use only individual count columns
            # 1. full data set
            X = df_shp_w_activities_socioecon_log[var_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon_log[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_full_log.loc[var_1, var_2] = r2
            
            # 2. data set without top K activities
            X = df_shp_w_activities_socioecon_without_max_log[var_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon_without_max_log[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_no_max_log.loc[var_1, var_2] = r2
            
            # 3. data set without top K activities and zero values
            X = df_shp_w_activities_socioecon_without_max_zeros_log[var_1]
            X = sm.add_constant(X)
            y = df_shp_w_activities_socioecon_without_max_zeros_log[var_2].values
            reg = LinearRegression().fit(X, y)
            r2 = np.round(reg.score(X, y), decimals = 3)
            r2_matrix_no_max_zeros_log.loc[var_1, var_2] = r2
            
```
