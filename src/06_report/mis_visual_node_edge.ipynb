# Visualizing nodes and edges in a schematic diagram

```python
import networkx as nx
import matplotlib.pyplot as plt
```

# 1 - Base node and edge visualization

```python
G = nx.Graph()
G.add_nodes_from([1,2,3,4,5,6])
pos = nx.shell_layout(G)
```

```python
plt.figure(figsize = (3,3))
nx.draw(G, pos = pos, node_color = 'grey', edge_color = 'red')
plt.show()
```

```python
G.add_edges_from([(1,2), (2,3), (2,5), (3,4), (3,5), (4,5)])
```

```python
plt.figure(figsize = (3,3))
nx.draw(G, pos = pos, node_color = 'grey', edge_color = 'red')
plt.show()
```

# 2 - Visualizing 2D grids

```python
G_grid = nx.grid_2d_graph(6, 6)  # 8x8 grid
pos = {(x,y):(y,-x) for x,y in G_grid.nodes()}
```

```python
G_two_nodes = nx.Graph()
G_two_nodes.add_nodes_from([(1,1), (2,4)])
pos_two_nodes = {(x,y):(y,-x) for x,y in G_two_nodes.nodes()}
```

```python
G_two_nodes_with_edge = nx.Graph()
G_two_nodes_with_edge.add_nodes_from([(1,1), (2,4)])
G_two_nodes_with_edge.add_edge((1,1), (2,4))
pos_two_nodes_with_edge = {(x,y):(y,-x) for x,y in G_two_nodes_with_edge.nodes()}
```

```python
plt.figure(figsize = (8,8))
nx.draw(G_grid, pos = pos, node_color = 'grey', edge_color = 'black', node_size = 800, width = 5)
# nx.draw(G_two_nodes, pos = pos, node_color = 'red')
plt.show()
```

```python
plt.figure(figsize = (8,8))
nx.draw(G_grid, pos = pos, node_color = 'grey', edge_color = 'black', node_size = 800, width = 5)
nx.draw(G_two_nodes, pos = pos, node_color = 'red', node_size = 800)
plt.show()
```

```python
plt.figure(figsize = (8,8))
nx.draw(G_grid, pos = pos, node_color = 'grey', edge_color = 'black', node_size = 800, width = 5)
nx.draw(G_two_nodes_with_edge, pos = pos, node_color = 'red', edge_color = 'red', node_size = 800, width = 2)
plt.show()
```
