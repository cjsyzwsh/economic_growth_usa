```python
### 
visit_counts = visit_counts.rename("# of visits")
unique_venues = unique_venues.rename("# of venues")
unique_users = unique_users.rename("# of users")
ave_duration = ave_duration.rename("Average duration")
ave_duration = np.round_(ave_duration, decimals = 1)
ave_weekday_ratio = ave_weekday_ratio.rename("Average weekday ratio")
ave_weekday_ratio = np.round_(ave_weekday_ratio, decimals = 3)
ave_home_distance = ave_home_distance.rename("Average home distance")
ave_home_distance = np.round_(ave_home_distance, decimals = 1)

activities = pd.concat([visit_counts, unique_venues, unique_users, ave_duration, ave_weekday_ratio, ave_home_distance], axis = 1)
sorted_activities = activities.sort_values('# of visits', ascending=False)

# add ranking
sorted_activities['Ranking of visits'] = np.arange(sorted_activities.shape[0]) + 1

sorted_activities
```

```python
# reporting strong mobility networks (top 100)
strong_ties = sorted_activities.iloc[:100, :]
strong_ties
```

```python
# reporting weak mobility networks (bottom 100)
weak_ties = sorted_activities.iloc[-100:, :]
weak_ties
```

```python
# union of the predictive activities
unique_predictive_activities = []
for city in ['boston','chicago','miami']:
    for output in ['inc_median_household_2018','property_value_median_2018','rent_median_2018']:
        unique_predictive_activities = unique_predictive_activities + list(coeff_dic[city][output]['lasso (no socio-demographics)'].index)
        
unique_predictive_activities = np.unique(unique_predictive_activities, return_counts=True)
unique_predictive_activities

# choose only those with more than 1 time. (For some robustness...)
unique_predictive_activities = unique_predictive_activities[0][unique_predictive_activities[1] > 1]
unique_predictive_activities
print(len(unique_predictive_activities))
```

```python
# reporting predictive mobility networks (union of all lasso coefficients)
# add a ranking column?
predictive_ties = sorted_activities.loc[unique_predictive_activities, :].sort_values('Ranking of visits', ascending=True)
predictive_ties
```
